```python
import pandas as pd

orders = pd.read_csv('data/orders.csv')
orders
```

```python
ads = pd.read_csv('data/ads.csv')
ads
```

```python
orders['dt'] = orders['created_at'].str.slice(0,10)
```

```python
daily_revenue = orders.groupby('dt').agg(revenue = ('payment_sum', 'sum'))
daily_revenue.head()
```

```python
daily_spend = ads.groupby('dt').agg(spend = ('ad_spend', 'sum'))
daily_spend.head()
```

```python
(
    daily_revenue
    .merge(daily_spend, how='outer', left_index=True, right_index=True)
    .fillna(0)
).head(30)
```
