```python
df = pd.read_csv('ads.csv')
df.head()
```

```python
# расходы на каждый рекламный канал в каждой стране
# стоимость клика в каждой стране и каждом рекламном канале
# данные были читаемыми
```

```python
df.pivot_table(
    index='country',
    columns='ad_channel',
    values=['ad_spend', 'clicks'],
    aggfunc='sum'
)
```

```python
df.set_index('country').loc['US'][['date', 'ad_channel', 'impressions']]
```

```python
(
    df
    .set_index('country')
    .loc['US']
    .pivot(
        index='date',
        columns='ad_channel',
        values='impressions'
    )
    .fillna(0)
    .astype(int)
)
```

```python
av_spend = df.groupby(['country', 'ad_channel'])['ad_spend'].sum().mean()

def get_av(values):
    x = sum(values)
    if x < av_spend:
        return 'below_av'
    elif x == av_spend:
        return 'av'
    else:
        return 'above_av'

df.pivot_table(
    index='country',
    columns='ad_channel',
    values='ad_spend',
    aggfunc=get_av
)

```
